# Importing RNA-seq data and formatting for analysis

```python
import pandas as pd
import numpy as np

# import sample sheet to sort between files and submitter ids
sampleIDs = pd.read_csv("~/code/Machine_Learning/Final_Project/gdc_sample_sheet.2019-05-12.tsv", sep='\t')

# add additional column for file location
sampleIDs['File_loc'] = 'rna_data/'+sampleIDs['File ID']+'/'+sampleIDs['File Name']

sampleIDs
```

## Create dataframe for solid tissue normal data

```python
# Subset data frame for only solid_tissue
STN = sampleIDs['Sample Type'] == 'Solid Tissue Normal'
Solid_Tissue = sampleIDs[STN]

# create iterable array for file locations
file_list_STN = Solid_Tissue['File_loc'].to_numpy()

# create iterable list for submitter ID to be assigned to imported data
sample_names_STN = Solid_Tissue['Case ID'].to_numpy()

# automate process of reading in files
li_STN = []

for file, sample in zip(file_list_STN, sample_names_STN):
    df = pd.read_csv(file, sep='\t', header =None, index_col=0, names=[sample])
    li_STN.append(df)
    
# merge all samples into single data frame
FPKM_STN = li_STN[0].join(li_STN[1:])
# transpose to samples are rows and genes are columns
FPKM_STN = FPKM_STN.T


FPKM_STN.head()
FPKM_STN.shape
```

```python
FPKM_STN.head()

```

## Create Dataframe for Primary Tumor samples

Note that some samples had multiple primary tumar expression data. This create problems in generating dataframes, so only the first sample was kept.

```python
# subset data
PT = sampleIDs['Sample Type'] == 'Primary Tumor'
Primary_Tumor = sampleIDs[PT]

# some repeated samples - get rid of these
Primary_Tumor = Primary_Tumor.drop_duplicates(subset = 'Case ID', keep = 'first')

# create iterable array for file locations
file_list_PT = Primary_Tumor['File_loc'].to_numpy()

# create iterable list for submitter ID to be assigned to imported data
sample_names_PT = Primary_Tumor['Case ID'].to_numpy()

# automate process of reading in files
li_PT = []

for file, sample in zip(file_list_PT, sample_names_PT):
    df = pd.read_csv(file, sep='\t', header =None, index_col=0, names=[sample])
    li_PT.append(df)
    
# merge all samples into single data frame
FPKM_PT = li_PT[0].join(li_PT[1:])
# transpose to samples are rows and genes are columns
FPKM_PT = FPKM_PT.T


FPKM_PT.head()
```

## Filtering out non-expressed genes

Any transcripts that did not have any reads (e.g. 0.0 FPKM values for all samples) were excluded.

```python
# Mark nonzero values as True (nonzero) or false (0.0)
zero_expr_STN = FPKM_STN.sum() != 0.0

FPKM_STN_nonzero = FPKM_STN.T[zero_expr_STN].T

FPKM_STN_nonzero.head()
```

```python
# Mark nonzero values as True (nonzero) or false (0.0)
zero_expr_PT = FPKM_PT.sum() != 0.0

FPKM_PT_nonzero = FPKM_PT.T[zero_expr_STN].T

FPKM_PT_nonzero.head()
```

```python
print(FPKM_STN_nonzero.shape)
print(FPKM_PT_nonzero.shape)
```

```python
# output configured data frames to CSV file
FPKM_STN_nonzero.to_csv('~/code/Machine_Learning/Final_Project/Solid_Tumor_FPKM.csv')
FPKM_PT_nonzero.to_csv('~/code/Machine_Learning/Final_Project/Primary_Tumor_FPKM.csv')
```
